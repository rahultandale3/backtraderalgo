```python

import pandas as pd
import math

def ohlc_to_renko(df, brick_size):
    renko_df = pd.DataFrame(columns=["date", "open", "close", "up", "down"])
    print(renko_df)
    print(df["close"].iloc[0])
    
    brick_value = brick_size 
    # brick_value = brick_size + df["close"].iloc[0]
    print(brick_value)
    
    up_brick = 0
    down_brick = 0
    up_list = [0]
    down_list = [0]
    # prev_price = df["close"].iloc[0]
    prev_price = math.floor(df["open"].iloc[0]/15)*15
    print(prev_price)
    
    prev_index = df.index[0]
    blank_list = []
    
    for i in range(1,len(df)):


        close_price = df["close"].iloc[i]
        diff_close_prev = close_price - prev_price
        blank_list.append(abs(diff_close_prev))
        max_num = max(blank_list)


        # global close_price
        # print(close_price)
        # print(prev_price)
        # prev_price = df["close"].iloc[i-1]
        date = df["date"].iloc[i]
        # print(date)
        if i ==1 :
          prev_price = math.floor(df["open"].iloc[i]/15)*15
          print(prev_price)
          if close_price - prev_price >= brick_value:
            up_brick += 1
            up_list.append(up_brick)
            down_list.append(0)
            close_price = prev_price +brick_value 
            renko_df.loc[len(renko_df)]= [date, prev_price, close_price, up_brick, 0]
            prev_price = close_price
            

          elif prev_price - close_price >= brick_value:
            down_brick += 1
            up_list.append(0)
            down_list.append(down_brick)
            close_price = prev_price - brick_value 
            renko_df.loc[len(renko_df)]= [date, prev_price, close_price, 0, down_brick]
            prev_price = close_price

          else:
            print("do nothing")




  
        if up_list[-1] != 0 : 
          if close_price - prev_price >= brick_value:
            n = math.floor((close_price - prev_price)/15)
            up_brick += 1
            up_list.append(up_brick)
            down_list.append(0)
            close_price = prev_price +brick_value*n
            renko_df.loc[len(renko_df)]= [date, prev_price, close_price, up_brick, 0]
            prev_price = close_price
            prev_index = date

          
          elif (prev_price-15 - close_price )>= brick_value:
            n = math.floor((prev_price-15 - close_price)/15) 
            down_brick += 1
            up_list.append(0)
            down_list.append(down_brick)
            close_price = prev_price-15 - brick_value*n
            prev_price = prev_price- 15 
            renko_df.loc[len(renko_df)]= [date, prev_price , close_price, 0, down_brick]
            prev_price = close_price
            prev_index = date

        elif down_list[-1] != 0 :
          if (close_price - prev_price-15 )>= brick_value:
            n = math.floor((close_price - prev_price-15)/15)
            up_brick += 1
            up_list.append(up_brick)
            down_list.append(0)
            close_price = prev_price + 15 +brick_value*n
            prev_price = prev_price + 15
            renko_df.loc[len(renko_df)]= [date, prev_price, close_price, up_brick, 0]
            prev_price = close_price
            prev_index = date

          elif prev_price- close_price >= brick_value:            
            n = math.floor((prev_price - close_price)/15) 
            down_brick += 1
            up_list.append(0)
            down_list.append(down_brick)
            close_price = prev_price - brick_value*n
            renko_df.loc[len(renko_df)]= [date, prev_price, close_price, 0, down_brick]
            prev_price = close_price
            prev_index = date
          
          
        elif down_list[-1] == 0 or up_list[-1] == 0 :
          print("we are in final cell")
          if close_price - prev_price >= brick_value:
            n = math.floor((close_price - prev_price)/15) 


            # print("u r in if ")previous cadle is not green then our previous price will be previous candle open 
            up_brick += 1
            up_list.append(up_brick)
            down_list.append(0)            
            # make code for if 
            close_price = prev_price +brick_value*n 
            renko_df.loc[len(renko_df)]= [date, prev_price, close_price, up_brick, 0]
            # renko_df.iloc[] = list
            prev_price = close_price
            prev_index = date
          elif prev_price - close_price >= brick_value:
            n = math.floor((prev_price - close_price)/15) 

            # print("u r in elif ")
            down_brick += 1
            up_list.append(0)
            down_list.append(down_brick)
            # renko_df.iloc[i] = [date, prev_price, close_price, 0, down_brick]
            # renko_df.loc[i]['date'] = date
            close_price = prev_price - brick_value*n
            renko_df.loc[len(renko_df)]= [date, prev_price, close_price, 0, down_brick]
            prev_price = close_price
            prev_index = date
          # else:
          #   print("u r in else ")
            # renko_df.loc[prev_index]["close"] = close_price
            # renko_df.loc[len(renko_df)]= [date, prev_price, close_price, up_brick, 0]
    print(renko_df.tail(10))       
    return renko_df ,blank_list , max_num

```
